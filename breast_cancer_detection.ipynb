**Building Model**

```python
X=data.drop(['diagnosis'],axis=1)
y = data['diagnosis']
a = df_model.drop(['diagnosis'],axis=1)
b = df_model['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=40)
a_train, a_test, b_train, b_test = train_test_split(a,b,test_size=0.3,random_state=40)
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

dtc=DecisionTreeClassifier(random_state=0)
model2=dtc.fit(X_train,y_train)
prediction2=model2.predict(X_test)
cm2= confusion_matrix(y_test,prediction2)
cm2

sns.set(font_scale=1.4) # set font size
sns.heatmap(cm2, annot=True, annot_kws={"size": 10}, cmap='Blues', cbar=False, fmt='g')

plt.suptitle("Decision Tree")
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

**Accuracy tells you how many times the ML model was correct overall. Precision is how good the model is at predicting a specific category. Recall tells you how many times the model was able to detect a specific category. The F-measure is calculated as the harmonic mean of precision and recall, giving each the same weighting**

**Decision Tree Hyperparameter Tuning using GridSearchCV**

```python
from scipy.stats.distributions import randint
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

params = {
    'max_depth': [5, 6, 7, 8, 9, None],
    'min_samples_leaf': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'criterion': ["gini", "entropy"]
}

model=DecisionTreeClassifier(random_state=99)

dtc_hyper = GridSearchCV(model,params,cv=5)
# dtc_hyper = RandomizedSearchCV(estimator = model, param_distributions = params,
#                n_iter = 90, cv = 5, verbose=2, random_state=35, n_jobs = -1)
dtc_hyper.fit(X_train,y_train.values.ravel())

def display(results):
    print(f'Best parameters are: {results.best_params_}')
    print("\n")
    mean_score = results.cv_results_['mean_test_score']
    std_score = results.cv_results_['std_test_score']
    params = results.cv_results_['params']
    for mean,std,params in zip(mean_score,std_score,params):
        print(f'{round(mean,3)} + or -{round(std,3)} for the {params}')
```

```python
dtc_hyper. best_params_ 
```

```python
#Predict testing set
y_pred_dtc = dtc_hyper.predict(X_test)
cm3=confusion_matrix(y_test,y_pred_dtc)

sns.set(font_scale=1.4) # set font size
sns.heatmap(cm3, annot=True, annot_kws={"size": 10}, cmap='Blues', cbar=False, fmt='g')

plt.suptitle("Decision Tree")
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.savefig('dt_cm_dpi_600.png', dpi=600)
plt.show()

#Check performance using accuracy
print(accuracy_score(y_test, y_pred_dtc))
#Check performance using ROC
print(roc_auc_score(y_test, y_pred_dtc))
#print
print(classification_report(y_test, y_pred_dtc))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
# plt.style.use('seaborn')

# Generate predictions on the test set
classifiers = {"Decision Tree": dtc_hyper, "Random Forest": cv, "KNN": best_model, "Naive Bayes": nbModel_grid, "Logistic Regression": best_lr}

# compute confusion matrix for each classifier
conf_matrices = []
for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    conf_matrices.append(confusion_matrix(y_test, y_pred))

# plot confusion matrices
fig, axes = plt.subplots(2, 3, figsize=(8, 8))
fig.suptitle("Confusion Matrices")

for i, (name, _) in enumerate(classifiers.items()):
    row = i // 3
    col = i % 3
    ax = axes[row, col]
    ax.imshow(conf_matrices[i], cmap=plt.cm.Blues)
    ax.set_xticks(np.arange(2))
    ax.set_yticks(np.arange(2))
    ax.set_xlabel("Predicted Label")
    ax.set_ylabel("True Label")
    ax.set_title(name)
    for j in range(2):
        for k in range(2):
            ax.text(k, j, str(conf_matrices[i][j, k]), ha="center", va="center", color="white" if conf_matrices[i][j, k] > conf_matrices[i].sum() / 2 else "black")

plt.tight_layout()
plt.show()
```
